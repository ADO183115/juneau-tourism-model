```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("simulation_results.csv")
df
```

```python
# Plot 1: T
plt.figure(figsize=(5, 3))
plt.plot(df["Year"], df["T"], color="red", marker="o", linewidth=2, label="Tourists T")
plt.xticks(df["Year"])

#Style
plt.title("Juneau Tourism Growth")
plt.xlabel("Year")
plt.ylabel("Number of Tourists T")
plt.grid(True)
plt.legend()
plt.show()

# Plot 2: E and Q   
plt.figure(figsize=(5, 4))
plt.plot(df["Year"], df["E"], color="green", marker="o", label="Environment Index E")
plt.plot(df["Year"], df["Q"], color="blue", marker="o", linewidth=2, label="Resident Satisfaction Q")
plt.ylim(0, 1.1)
plt.xticks(df["Year"])

# Stule
plt.title("Change in E and Q")
plt.xlabel("Year")
plt.ylabel("Index Values")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Layer 1: Government Reinvestment M
plt.bar(df["Year"], df["M"], color="green", label="Tourism Reinvestment M")

# Layer 2: Government Tax Revenue G-M
plt.bar(df["Year"], df["G"] - df["M"], bottom=df["M"], color="red", label="Government Tax Revenue G - M")

# Layer 3: Private Revenue from Tourism R-G
plt.bar(df["Year"], df["R"] - df["G"], bottom=df["G"], color="blue", label="Private Revenue R - G")

# Style
plt.title("Total Tourism Revenue R Breakdown")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Amount ($)", fontsize=12)
plt.xticks(df["Year"]) # Keep those whole numbers!
plt.legend(loc="upper left")
plt.grid(axis="y", linestyle="--", alpha=0.3)

plt.show()
```
